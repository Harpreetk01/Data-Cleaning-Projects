## Data preprocessing and cleaning 

```python
#Dropping not useful columns

data.drop(['index','Founded', 'Competitors'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
#Drop Duplicate

len(data[data.duplicated()])
```

```python
data.drop_duplicates(inplace = True)
```

Droped duplicate values from the data

```python
data.shape
```

```python
#Company Name Column 

data['Company Name'] = data['Company Name'].apply(lambda x:x.split("\n")[0])
```

```python
data.head(3)
```

Cleaning Company name column by removing Numbers and string values from company name

```python
data['Salary Estimate'].value_counts()
```

```python
data.head()
```

```python
len(data)
```

```python
#Replacing String words from the Salary Column

data['Salary Estimate'] = data['Salary Estimate'].str.replace("$", "", regex=True).str.replace("K", '', regex=True).str.replace("(Glassdoor est.)","", regex=True).str.replace("(Employer est.)","", regex=True)
```

```python
data['Salary Estimate'] = data['Salary Estimate'].str.replace(r"\(", "", regex=True).str.replace(r"\)", "", regex=True)
```

```python
data['min_salary'], data['max_salary'] = data['Salary Estimate'].str.split("-").str

data['min_salary'] = data['min_salary'].astype(float)
data['max_salary'] = data['max_salary'].astype(float)

data['avg_salary'] = (data['min_salary'] + data['max_salary']) / 2
```

```python
data.head()
```

```python
data.info()
```

```python
#Replace string values from size columns 

data['Size'] = data['Size'].str.replace("employees", "").str.replace('to', '-')
```

```python
#Rename the Size Column

data.rename(columns = {'Size': 'Employees_Size'}).head()
```

```python
#Cleaning Job Description column 

data['Job Description'] = data['Job Description'].apply(lambda x:x.split("\n"))
```

```python
data['Job Description'].head(10)
```

```python
#Creating new column named - job_state

data['job_state'] = data['Location'].str.split(', ').str[1]
```

```python
data.head()
```

```python
#Creating New Column named - Seniority 

def seniority(job_title):
    job_title=job_title.lower()
    snr=['sr','senior','lead','principal','vp','vice president','director']
    for i in snr:
        if i in job_title:
            return "senior"
    if "jr" in job_title:
        return "junior"
    return "na"
```

```python
data['Seniority'] = data['Job Title'].apply(seniority)
```
